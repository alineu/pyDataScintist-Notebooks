# `read_csv()`: read data from csv and tsv files

```python
import pandas as pd
dataPath = 'https://raw.githubusercontent.com/alineu/pyDataScintist-Notebooks/main/data/'
df = pd.read_csv(dataPath+'BostonHousing_min.csv')
df.head()
```

### `sep`: Delimiter to use - the default value is ','

```python
# Not specifying the delimiter when reading a tsv file is problematic
pd.read_csv(dataPath+'BostonHousing_min.tsv').head()
```

```python
pd.read_csv(dataPath+'BostonHousing_min.tsv', 
            sep='\t').head()
```

### `header`: specifies what row to start reading the csv - the deafult value is 0

```python
# header=0 is the default behavior
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            header=0).head()
```

```python
# header=n where n is an integer reads the file starting at the nth line
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            header=3).head()
```

`header=None` treats the first line of the csv file as data and the column names will be assigned as a list of integers

```python
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            header=None).head()
```

```python
# So header=None is convenient if the file does not have a header
pd.read_csv(dataPath+'BostonHousing_no-header_min.csv', 
            header=None).head()
```

### `prefix` argument can be specified to be added to column numbers when no header

```python
pd.read_csv(dataPath+'BostonHousing_no-header_min.csv', 
            header=None, 
            prefix='col_').head()
```

### `skiprows` argument can be used to define:
- line numbers to skip (list) e.g. `skiprows=[1, 5, 8]`
- number of lines to skip (int) at the start of the filee.g. `skiprows=5` (disregards the header as well)
- a callable function which is evaluated against the row indices e.g. `skiprows=lambda x: x%3==2`

```python
pd.read_csv(dataPath+'BostonHousing.csv', 
            skiprows=range(2,20)).head()
```

```python
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            skiprows=lambda x: x%3==2).head()
```

### `names` : change/assign the column names:

```python
col_names = [f'col_{i}' for i in range(len(df.columns))]
col_names
```

```python
pd.read_csv('https://raw.githubusercontent.com/alineu/pyDataScintist-Notebooks/main/data/BostonHousing.csv', 
            names=col_names, 
            header=0).head()
```

### `usecols` - include specific columns in the DataFrame

```python
### using column names
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            usecols=['age', 'tax']).head()
```

```python
# Using column ids
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            usecols=[0, 5]).head()
```

### `nrows` - read pieces of a large csv

```python
pd.read_csv(dataPath+'BostonHousing_min.csv', 
            nrows=2, 
            usecols=[0, 5]).head()
```

### `dtype`: specify the column data type

```python
df = pd.read_csv(dataPath+'BostonHousing_min.csv', 
                 usecols=['age', 'tax'])
df.dtypes
```

```python
df.head()
```

```python
# Change the "tax" column data type from int to float
df = pd.read_csv(dataPath+'BostonHousing_min.csv',  
            usecols=['age', 'tax'], 
            dtype={'tax': float})
df.dtypes
```

```python
# Note the floating point in the "tax" column
df.head()
```
